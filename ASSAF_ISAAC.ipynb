##Évolution de la Consommation de Gaz et de Charbon aux États-Unis (1999-2023)

Depuis la fin des années 90, les États-Unis ont connu une transformation remarquable dans leur mix énergétique, marquée par une augmentation significative de la consommation de gaz naturel au détriment du charbon. Cette évolution coïncide avec la révolution des schistes, un tournant majeur dans la production de gaz naturel américain qui a bouleversé non seulement le marché énergétique intérieur mais aussi le marché mondial.

La consommation de charbon, autrefois prédominante dans la production d'électricité, a été progressivement remplacée par le gaz naturel, perçu comme une alternative moins polluante et plus économique. Cette transition a été encouragée par des politiques environnementales visant à réduire les émissions de gaz à effet de serre, dans un contexte de préoccupations croissantes liées au changement climatique.

Cependant, cette transition soulève des questions importantes sur la durabilité à long terme de la dépendance au gaz naturel, en particulier en ce qui concerne les émissions de méthane, les risques liés à l'exploitation des gaz de schiste et la volatilité des prix. De plus, les implications pour l'industrie du charbon, autrefois pilier de l'économie dans certaines régions, et pour les communautés qui en dépendent, sont également des sujets de préoccupation.

####La révolution du gaz de schiste au US
La révolution du gaz de schiste aux États-Unis, qui a pris son essor au milieu des années 2000, est avant tout une prouesse technologique. Cette révolution repose sur deux innovations clés : la fracturation hydraulique et le forage horizontal, qui ont permis d'accéder aux vastes réserves de gaz naturel piégées dans les formations de schiste.

La fracturation hydraulique, ou "fracking", est une technique qui consiste à injecter un mélange d'eau, de sable et de produits chimiques sous haute pression dans les formations rocheuses de schiste. Cette pression crée des fissures dans la roche, libérant ainsi le gaz naturel emprisonné. Cette méthode a révolutionné la capacité à exploiter des réserves de gaz autrefois inaccessibles.

Complémentaire à la fracturation hydraulique, le forage horizontal permet d'atteindre et de libérer le gaz de schiste de manière plus efficace et étendue. Contrairement au forage traditionnel vertical, le forage horizontal peut s'étendre sur des kilomètres à partir du puits vertical, augmentant ainsi considérablement la surface de roche accessible pour la fracturation.

Le développement rapide de ces technologies a permis aux États-Unis de devenir l'un des plus grands producteurs de gaz naturel au monde. La capacité à exploiter le gaz de schiste a non seulement augmenté la production de gaz naturel, mais a également contribué à diversifier les sources d'énergie du pays.

Bien que la révolution du gaz de schiste soit une réussite technique, elle soulève également des enjeux environnementaux, notamment en matière de gestion de l'eau, de risques de tremblements de terre induits et de fuites de méthane. Ces défis nécessitent une attention continue et des solutions innovantes pour assurer une exploitation responsable.

###Les conséquences de cette transition sur les émissions de C02

Selon les données de l'Energy Information Administration (EIA), le secteur de la production d'électricité aux États-Unis a connu une baisse de 32% de ses émissions de CO2 entre 2005 et 2019. En 2019, les émissions de CO2 du secteur électrique étaient de 1,724 millions de tonnes métriques, contre 2,544 millions de tonnes métriques en 2005. Cette réduction est largement attribuée à la transition du charbon au gaz naturel pour la production d'électricité. En 2005, le charbon représentait 50% de la production d'électricité américaine, part qui a chuté à 23% en 2019, tandis que la part du gaz naturel est passée de 19% à 38% sur la même période

Lorsqu'il est utilisé pour générer de l'électricité, le gaz naturel émet nettement moins de CO2 que le charbon. En 2019, la production d'électricité à partir du charbon a émis 2,257 livres de CO2 par mégawattheure (MWh), tandis que celle du gaz naturel a produit moins de la moitié de ce montant, à savoir 976 livres de CO2/MWh. Cela s'explique par le fait que le charbon a une teneur en carbone plus élevée par unité d'énergie par rapport au gaz naturel. De plus, les centrales à gaz sont généralement plus efficaces que les centrales à charbon, ce qui contribue également à la réduction des émissions de CO2

###Les limites de la transition du charbon au gaz naturel :
La substitution initiale du charbon par le gaz a entraîné des réductions significatives de CO2, car le gaz émet environ la moitié du CO2 du charbon lors de la combustion. Cependant, cette efficacité diminue à mesure que les centrales à charbon les moins économiques sont déjà remplacées ou mises hors service, laissant celles plus efficaces et moins susceptibles d'être remplacées par du gaz.

Les centrales à gaz, bien que plus propres que celles à charbon, ont des limites en termes d'efficacité énergétique. La réduction supplémentaire des émissions dépendra de l'amélioration de l'efficacité de ces centrales ou de la transition vers des sources d'énergie renouvelables.

Les fuites de méthane lors de l'extraction et du transport du gaz naturel sont un problème majeur. Le méthane est un gaz à effet de serre beaucoup plus puissant que le CO2 sur le court terme, et ces fuites pourraient annuler une partie des bénéfices obtenus en réduisant la combustion du charbon.

##ARMA


On choisi d'analyser la consommation différenciée du gaz :


```r
acf(cgaz_diff)
pacf(cgaz_diff)
```

###Algorithme d'optimisation du paramétrage de l'ARMA

```r
p_max <- 5
q_max <- 5

# Créer un dataframe pour stocker les résultats
results <- data.frame(p=integer(), q=integer(), AIC=double(), BIC=double())

for (p in 0:p_max) {
  for (q in 0:q_max) {
    # Ajuster le modèle ARMA
    model <- tryCatch(Arima(cgaz_diff, order=c(p, 0, q)), error=function(e) NULL)

    # Vérifier si le modèle est non-nul et a réussi à converger
    if (!is.null(model)) {
      current_aic <- AIC(model)
      current_bic <- BIC(model)
      results <- rbind(results, data.frame(p=p, q=q, AIC=current_aic, BIC=current_bic))
    }
  }
}
results
```

Il semblerait que le ARMA(2,4) est le meilleur choix.


###Test des coefficients

```r
ARMA24<-Arima(y = cgaz_diff,order=c(2,0,4))

coeftest(ARMA24)
autoplot(ARMA24)
```

En valeur absolue tout les coefficiens ont une valeur inférieure à 1, il n'y a donc pas de racine unitaire. Il est important de souligner que les coefficients sont très significatifs.

###Test sur les résidus

####Ljung-Box test

```r
test_res<-checkresiduals(ARMA24,lag = 15)
```

La statistique de test (Q*) est significativement élevée. La p-value est extrêmement faible (< 2.2e-16). On rejette donc l'hypothèse nulle selon laquelle les résidus du modèle ARIMA(2,4) ne présentent pas d'autocorrélation jusqu'au lag 15.

####Test de normalité des résidus

```r
which(ARMA24$residuals==max(ARMA24$residuals))
which(ARMA24$residuals==min(ARMA24$residuals))
```

```r
gghistogram(ARMA24$residuals[-250], add.normal = TRUE)

sprintf("Skewness des residus : %f",skewness(ARMA24$residuals[-250]))
sprintf("kurtosis des residus : %f",kurtosis(ARMA24$residuals[-250]))
qqnorm(ARMA24$residuals[-250])
qqline(ARMA24$residuals[-250])
```

# PARTIE II : MODÉLISATION MULTIVARIÉE

##VAR

```r
plot(ccoal_diff,cgaz_diff)
```

Choix du nombre de lag

```r
data_VAR <- data.frame(ccoal_diff, cgaz_diff)
K_select <-VARselect(y=data_VAR,lag.max = 12,type = "none" )
K_select$selection
```

On choisi 12 retards

```r
var_model <- VAR(data_VAR, p = 12, type ="none" )
summary(var_model)
plot(var_model)
```

L'erreur standard des résidus pour la consommation de charbon est de 3.47 et le R² ajusté est de 0.7934, tandis que pour la consommation de gaz, l'erreur standard des résidus est de 3.598 et le R² ajusté est de 0.8502, ce qui suggère un meilleur ajustement du modèle pour la consommation de gaz.

La matrice de covariance des résidus révèle une corrélation de 0.5341 entre les erreurs des deux équations, indiquant une corrélation modérée entre les chocs affectant les consommations de charbon et de gaz.

En somme, les résultats indiquent une interdépendance entre les consommations de charbon et de gaz, avec des dynamiques complexes où les données passées influencent les variations futures, et le modèle semble capturer efficacement ces relations.

##Test de causalité


```r
causality_test <- causality(var_model, cause=c('ccoal_diff'))
causality_test
```

 les résultats de ces tests suggèrent non seulement que les variations passées dans la consommation de charbon ont un pouvoir prédictif sur les variations futures de la consommation de gaz (causalité de Granger), mais aussi que les variations entre ces deux séries sont simultanément interdépendantes (causalité instantanée). Cela implique que des chocs affectant la consommation de charbon peuvent influencer la consommation de gaz et vice versa, presque instantanément.

##Impulsion réponse


```r
irf_results <- irf(var_model, n.ahead = 15, boot = TRUE, ci = 0.95)
plot(irf_results)
```

###Méthode des projections local

```r
irf <- lp_lin(endog_data =data_VAR,lags_criterion = 'AIC',max_lags = 12,trend = 0,hor = 15,shock_type = 0,confint = 1.96,lags_endog_lin = NaN)
plot(irf)
```

Les deux variables présentent une réactivité aux chocs réciproques ainsi qu'aux chocs internes, indiquant une autocorrélation. Il est notable que la réaction de la consommation de charbon à un choc sur la consommation de gaz paraît moindre comparativement à la réponse plus marquée de la consommation de gaz à un choc sur la consommation de charbon.

##Test de Johansen

```r
test_johansen <- ca.jo(data_VAR, type="trace", ecdet="none",spec = "transitory", K=12)
summary(test_johansen)
```

Le test de Johansen indique au moins une relation de cointégration entre la consommation de charbon et de gaz, avec des statistiques de test dépassant largement les valeurs critiques. Cela suggère une relation d'équilibre à long terme entre ces deux variables.

Test de la valeur propre maximal

```r
test_johansen2 <- ca.jo(data_VAR, type="eigen", ecdet="none",spec = "transitory", K=12)
summary(test_johansen2)
```

Le test de cointégration de Johansen, utilisant la statistique de la plus grande valeur propre , montre une forte présomption de cointégration entre la consommation de charbon et de gaz. La statistique de test pour r = 0 (pas de relation de cointégration) est de 79.58, dépassant les valeurs critiques à tous les niveaux de confiance, indiquant au moins une relation de cointégration. Ces résultats confirment une relation entre la cosommation de charbon et de gaz sur le long terme.

Les modèles montrent que les variations passées dans la consommation de charbon et de gaz affectent leur consommation actuelle. Les résultats soulignent une forte interconnexion entre ces marchés énergétiques.
